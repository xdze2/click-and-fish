```python
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 102)
```

```python
x_grid, y_grid = np.meshgrid(x, y)

r_grid = np.sqrt( x_grid**2 + y_grid**2 )
theta_grid = np.arctan2(y_grid, x_grid)
E = np.cos( np.pi/2* np.cos(theta_grid) )/np.sin(theta_grid)/r_grid
```

```python
E[r_grid<0.2] = 0
```

```python
#plt.pcolormesh(E)
plt.contour(np.log(np.abs(E)))
plt.axis('square')
```
